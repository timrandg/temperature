```python
from matplotlib import cm
import joypy
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

temp = pd.read_csv("TAVG_daily.txt", delim_whitespace=True ,comment="%", header=0)
```

```python
labels=[y if y%10==0 else None for y in list(temp.Year.unique())]
```
